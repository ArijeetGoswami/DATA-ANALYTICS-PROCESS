```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/content/categories.csv")
df2 = pd.read_csv("/content/groups.csv")
```

```python
df1
```

```python
df2
```

```python
#lets create a pie chart for the following df1
pie = df1.groupby(['Unnamed: 0']).sum().plot(kind = 'pie', subplots=True , figsize = (10,8))
```

```python
plt.figure(figsize=(10, 6))

# Create a bar chart
plt.barh(df2['CATEGORIES'],df2['COST'])

# Add titles and labels (optional)
plt.title('Cost by Category')
plt.xlabel('Cost')
plt.ylabel('Categories')

# Show the plot
plt.show()
```

```python
#lets also have a check it with a pie chart
pie2 = df2.groupby(['CATEGORIES']).sum().plot(kind='pie',subplots=True,figsize=(20,20))
```

```python
#lets have a histo-gram for the df2 data-set
plt.figure(figsize=(10,6))
plt.hist(df2['CATEGORIES'],edgecolor='white', orientation='horizontal')
plt.xlabel('frequency')
plt.ylabel('categories')
plt.title('fequency of items')
plt.show()
```
